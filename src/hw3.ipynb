```python
import os

def align_rnaseq(file):
  name = os.path.splitext(file)[0]
  if not os.path.isfile(name+".fastq"):
    download_rnaseq([name.split('_')[0]])

  !time hisat2 -p 3 -x mm10/genome -U {name}.fastq -S {name}.sam 2>{name}.hisat
  !rm {name}.fastq
  !grep -P '^@|NH:i:1$' {name}.sam > {name}.uniq.sam
  !rm {name}.sam
  !time htseq-count --format=sam --stranded=no {name}.uniq.sam gencode.vM25.annotation.gtf > {name}.counts
  !rm {name}.uniq.sam
  !grep '^__' {name}.counts #?
```

```python
align_rnaseq("SRR3414629_1.fastq")
align_rnaseq("SRR3414630_1.fastq")
align_rnaseq("SRR3414631_1.fastq")
align_rnaseq("SRR3414635_1.fastq")
align_rnaseq("SRR3414636_1.fastq")
align_rnaseq("SRR3414637_1.fastq")
```
